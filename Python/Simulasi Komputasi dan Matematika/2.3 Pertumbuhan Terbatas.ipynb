```
"""
Graph of logistic equation, where initial population is 20, carrying capacity is
1000, and instantaneous rate of change of births is 50%, with time (t) in years
"""



## Kamus:
## INPUT
simulationLength = int(input("Masukkan lama simulasi: "))
population = int(input("Masukkan jumlah populasi awal: "))
carrying_capacity = int(input("Masukkan kapasitas lingkungan: "))
growthRate = float(input("Masukkan konstanta kelahiran populasi: "))
delta_t = float(input("Masukkan selang waktu: "))
numIterations = int(simulationLength/delta_t)
populasi = []
time = []

for i in range(0,numIterations):
  growth = (growthRate*population) - growthRate*(population/carrying_capacity)*population
  t = i*delta_t
  #print(t,"\t",growth,"\t",population)
  population = population + growth*delta_t
  populasi.append(population)
  time.append(t)

import matplotlib.pyplot as plt
plt.plot(time,populasi)
plt.xlabel("Time(hours)")
plt.ylabel("Population ")
plt.show()
```
