```
"""
                    GERAK JATUH
"""
# INPUT
position = float(input("Masukkan jarak dari tanah = ")) # 400
radius = 0.05 
projected_area = 3.14159 * radius**2
#projected_area = float(input("Masukkan luas daerah proyeksi = ")) # 0.1
mass = float(input("Masukkan massa  = ")) # 0.5
velocity = 0 #m/s
acceleration = -9.81 #m/s2
weight = mass * acceleration
simulationLength = int(input("Masukkan lama simulasi: "))
delta_t = float(input("Masukkan selang waktu: "))
numIterations = int(simulationLength/delta_t)

speed = []
positions = []
time = []

for i in range(0,numIterations+1):
  air_friction = -0.65 * projected_area * velocity * abs(velocity)
  total_force = weight + air_friction
  acceleration = total_force/mass
  velocity = velocity + acceleration*delta_t
  velocitys =abs(velocity)
  position = position + velocity * delta_t
  t = i*delta_t
  speed.append(velocitys)
  positions.append(position)
  time.append(t)

import matplotlib.pyplot as plt
plt.plot(time,positions,label="Position")
plt.plot(time,speed,label="Speed")
plt.xlabel("Time(hours)")
plt.legend()
plt.show()
```
