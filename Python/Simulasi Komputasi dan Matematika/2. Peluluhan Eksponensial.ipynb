```
"""
Laju perubahan massa suatu zat radiaktif sebanding dengan massa zat
tersebut.
Konstanta perubahan merupakan bilangan negatif.
Contohnya untuk Carbon-14, konstanta perubahannya ‚àí0.000120968.
Maka persamaan diferensial untuk masalah ini adalah
ùëëùëÑ/dt = ‚àí0.000120968 ùëÑ
dengan solusi ùëÑ = ùëÑ_0*ùëí**(‚àí0.000120968*ùë°)
Prinsip ini yang digunakan untuk menentukan usia fosil berdasarkan
kandungan Carbon-14 di dalam fosil tersebut (Carbon dating).
"""

# Model Persamaan beda peluluhan zat radiaktif di atas yaitu 
# Q(t) = Q(t-delta_t) - 0.000120968*Q(t-delta_t)*delta_t
## INPUT dengan MODEL MALTHUS I
simulationLength = int(input("Masukkan lama simulasi: "))
massa = float(input("Masukkan besar massa Carbon-14: "))
growthRate = 0.000120968
delta_t = float(input("Masukkan selang waktu: "))
numIterations = int(simulationLength/delta_t)
mass = []
time = []

for i in range(0,numIterations):
  growth = growthRate*massa
  t = i*delta_t
  #print(t,"\t",growth,"\t",population)
  massa = massa - growth*delta_t
  mass.append(massa)
  time.append(t)

import matplotlib.pyplot as plt
plt.plot(time,mass)
plt.xlabel("Time(hours)")
plt.ylabel("Massa Carbon-14")
plt.show()
```
