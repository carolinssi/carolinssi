```
"""
Banyaknya darah dalam tubuh orang dewasa 5 liter.
Banyaknya plasma (cairan yang memuat sel darah) sekitar 3 liter.
Untuk orang dewasa, dosis untuk mengobati sakit kepala adalah 1
atau 2 tablet dengan berat 325mg setiap 4 jam, maksimum 12
tablet per hari.
Penghilang rasa sakit akan efektif pada level 150 sampai 300
(μg/mL), sementara keracunan dapat terjadi pada konsentrasi
plasma 350 μg/mL.
Waktu paruh dari dosis 300 sampai 650 mg adalah 3.1 sampai 3.2
jam, dengan dosis yang lebih banyak memiliki waktu paruh yang
lebih lama.
"""


simulationLength = int(input("Masukkan lama simulasi: "))
half_life = float(input("Masukkan waktu paruh:"))
plasma_volume = float(input("Masukkan volume plasma: "))
aspirin = int(input("Masukkan jumlah aspirin awal: "))
elimination_constant = 0.6931471805599453/half_life
delta_t = float(input("Masukkan selang waktu: "))
numIterations = int(simulationLength/delta_t)
aspirin_concentration = []
time = []

for i in range(0,numIterations+1):
  plasma_concentration = aspirin /plasma_volume
  elimination = (elimination_constant*aspirin) 
  t = i*delta_t
  aspirin = aspirin - elimination*delta_t
  aspirin_concentration.append(plasma_concentration)
  time.append(t)

import matplotlib.pyplot as plt
plt.plot(time,aspirin_concentration)
plt.xlabel("Time(hours)")
plt.ylabel("Plasma Concentration ")
plt.show()
```
