```
"""
Algortima Model Malthus I
initialize 𝑠𝑖𝑚𝑢𝑙𝑎𝑡𝑖𝑜𝑛𝐿𝑒𝑛𝑔𝑡ℎ, 𝑝𝑜𝑝𝑢𝑙𝑎𝑡𝑖𝑜𝑛, 𝑔𝑟𝑜𝑤𝑡ℎ𝑅𝑎𝑡𝑒, ∆𝑡
𝑛𝑢𝑚𝐼𝑡𝑒𝑟𝑎𝑡𝑖𝑜𝑛𝑠 ← 𝑠𝑖𝑚𝑢𝑙𝑎𝑡𝑖𝑜𝑛𝐿𝑒𝑛𝑔𝑡ℎ / ∆𝑡
for 𝑖 from 0 to 𝑛𝑢𝑚𝐼𝑡𝑒𝑟𝑎𝑡𝑖𝑜𝑛𝑠 do:
  𝑔𝑟𝑜𝑤𝑡ℎ ← 𝑔𝑟𝑜𝑤𝑡ℎ𝑅𝑎𝑡𝑒 ∗ 𝑝𝑜𝑝𝑢𝑙𝑎𝑡𝑖𝑜𝑛
  𝑝𝑜𝑝𝑢𝑙𝑎𝑡𝑖𝑜𝑛 ← 𝑝𝑜𝑝𝑢𝑙𝑎𝑡𝑖𝑜𝑛 + 𝑔𝑟𝑜𝑤𝑡ℎ ∗ ∆𝑡
  𝑡 ← 𝑖 ∗ ∆𝑡
  display 𝑡, 𝑔𝑟𝑜𝑤𝑡ℎ, 𝑝𝑜𝑝𝑢𝑙𝑎𝑡𝑖𝑜n
"""

"""
We started with a bacterial population of size 100, an instantaneous growth rate of 10% = 0.10, and time
measured in hours. The Time Step is 0.005 hours. 
"""
## Kamus:
## INPUT
simulationLength = int(input("Masukkan lama simulasi: "))
population = int(input("Masukkan jumlah populasi awal: "))
growthRate = float(input("Masukkan konstanta pertumbuhan populasi: "))
delta_t = float(input("Masukkan selang waktu: "))
numIterations = int(simulationLength/delta_t)
populasi = []
time = []

for i in range(0,numIterations):
  growth = growthRate*population
  t = i*delta_t
  #print(t,"\t",growth,"\t",population)
  population = population + growth*delta_t
  populasi.append(population)
  time.append(t)

import matplotlib.pyplot as plt
plt.plot(time,populasi)
plt.xlabel("Time(hours)")
plt.ylabel("Population ")
plt.show()
```
