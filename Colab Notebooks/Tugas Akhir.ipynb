#**Perbedaan Fungsi**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan fungsi f(x, y, z) dan g(x, y, z)
def f(x, y, z, a):
    return a * x * y * z / ((1 + x + z) * (1 + y))

def g(x, y, z, a):
    return a * x * y * z / (1 + x + y + z)

# Rentang nilai x, y, z
x_vals = np.linspace(0, 1, 100)
y_vals = np.linspace(0, 1, 100)
z_vals = np.linspace(0, 1, 100)

# Inisialisasi array untuk menyimpan nilai perbedaan
diff_vals = np.zeros((100, 100, 100))

# Loop untuk menghitung nilai perbedaan pada setiap titik
for i in range(len(x_vals)):
    for j in range(len(y_vals)):
        for k in range(len(z_vals)):
            x = x_vals[i]
            y = y_vals[j]
            z = z_vals[k]
            a = 0.5  # Ubah nilai a sesuai keinginan

            diff_vals[i, j, k] = f(x, y, z, a) - g(x, y, z, a)

# Ambil nilai perbedaan pada titik (x=1/3, y=1/3, z=1/3)
x_index = np.argmin(np.abs(x_vals - 1/3))
y_index = np.argmin(np.abs(y_vals - 1/3))
z_index = np.argmin(np.abs(z_vals - 1/3))
perbedaan_1_3 = diff_vals[x_index, y_index, z_index]

# Plot kurva perbedaan pada titik (x=1/3, y=1/3, z=1/3)
plt.plot(x_vals, diff_vals[:, y_index, z_index], label='Perbedaan pada x=1/3')
plt.axhline(y=perbedaan_1_3, color='red', linestyle='--', label='Nilai perbedaan pada x=1/3, y=1/3, z=1/3')
plt.xlabel('Nilai x')
plt.ylabel('Perbedaan')
plt.legend()
plt.grid(True)
plt.title('Kurva Perbedaan Antara f(x, y, z) dan g(x, y, z) pada x=1/3, y=1/3, z=1/3')
plt.show()
```

#**Fungsi interaksi Holling Monolingual dengan bilingual**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan fungsi f(x, z) dengan a sebagai konstanta
def f(a, x, z):
    return a * x * z / (1 + x + z)

# Rentang nilai x, z
x_vals = np.linspace(0, 1, 100)
z_vals = np.linspace(0, 1, 100)

# Inisialisasi array untuk menyimpan nilai fungsi
result_vals = np.zeros((100, 100))

# Nilai konstanta a
a = 0.3

# Loop untuk menghitung nilai fungsi pada setiap titik
for i in range(len(x_vals)):
    for j in range(len(z_vals)):
        x = x_vals[i]
        z = z_vals[j]

        # Berdasarkan batasan x + z = 1
        if x + z <= 1:
            result_vals[i, j] = f(a, x, z)
        else:
            result_vals[i, j] = np.nan

# Plot grafik 2D pada bidang xz
X, Z = np.meshgrid(x_vals, z_vals)
plt.contourf(X, Z, result_vals, levels=50, cmap='viridis')
plt.colorbar()
plt.xlabel('Nilai $x_1$')
plt.ylabel('Nilai $x_3$')
plt.title('Grafik Fungsi f($x_1$, $x_3$) dengan a=0.3 pada $x_1$+$x_3$=1')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan fungsi f(x, z) dengan a sebagai konstanta
def f(a, x, z):
    return a * x * z / (1 + x + z)

# Rentang nilai x, z
x_vals = np.linspace(0, 1, 100)
z_vals = np.linspace(0, 1, 100)

# Inisialisasi array untuk menyimpan nilai fungsi
result_vals = np.zeros((100, 100))

# Nilai konstanta a (dalam range 0<=a<=1)
a_vals = np.linspace(0, 1, 10)

# Loop untuk menghitung nilai fungsi pada setiap titik dan setiap nilai a
for a in a_vals:
    for i in range(len(x_vals)):
        for j in range(len(z_vals)):
            x = x_vals[i]
            z = z_vals[j]

            # Berdasarkan batasan x + z = 1
            if x + z <= 1:
                result_vals[i, j] = f(a, x, z)
            else:
                result_vals[i, j] = np.nan

    # Plot grafik fungsi f(x, z) untuk setiap nilai a
    plt.plot(x_vals, result_vals[:, 50], label=f'a = {a:.2f}')

# Tambahkan legend dan label pada sumbu
plt.legend()
plt.xlabel('Nilai x')
plt.ylabel('f(x, 1-x)')
plt.title('Grafik Fungsi f(x, z) pada z=1-x dengan variasi a')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definisikan fungsi f1(x, z) dan f2(y, z)
def f1(a, x, z):
    return a * x * z / (1 + x + z)

def f2(b, y, z):
    return b * y * z / (1 + y + z)

# Inisialisasi array untuk menyimpan nilai fungsi f1(x, z) dan f2(y, z)
x_vals = np.linspace(0, 1, 100)
z_vals = np.linspace(0, 1, 100)

X, Z = np.meshgrid(x_vals, z_vals)

# Inisialisasi nilai konstanta a dan b
a = 0.7
b = 0.3

# Hitung nilai fungsi f1(x, z) dan f2(y, z)
F1 = f1(a, X, Z)
F2 = f2(b, 1 - X, Z)  # karena x + y + z = 1, maka y = 1 - x

# Hitung perbedaan fungsi f1(x, z) dan f2(y, z)
diff = F1 - F2

# Membuat plot 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot kurva perbedaan fungsi dalam 3D
ax.plot_surface(X, Z, diff, cmap='viridis')

# Tambahkan label sumbu dan judul
ax.set_xlabel('Nilai x')
ax.set_ylabel('Nilai z')
ax.set_zlabel('Perbedaan fungsi')
ax.set_title('Grafik 3D Perbedaan Fungsi f1(x, z) dan f2(y, z)')

# Tampilkan grafik
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definisikan fungsi f(x, y, z) dengan a sebagai konstanta
def f(a, x, y, z):
    return a * x * y * z / ((1 + x + z) * (1 + y))

# Rentang nilai x, y, z
x_vals = np.linspace(0, 1, 100)
y_vals = np.linspace(0, 1, 100)
z_vals = np.linspace(0, 1, 100)

# Inisialisasi array untuk menyimpan nilai fungsi
result_vals = np.zeros((100, 100, 100))

# Nilai konstanta a
a = 0.3

# Loop untuk menghitung nilai fungsi pada setiap titik
for i in range(len(x_vals)):
    for j in range(len(y_vals)):
        for k in range(len(z_vals)):
            x = x_vals[i]
            y = y_vals[j]
            z = z_vals[k]

            # Berdasarkan batasan x + y + z = 1
            if x + y + z <= 1:
                result_vals[i, j, k] = f(a, x, y, z)
            else:
                result_vals[i, j, k] = np.nan

# Plot grafik 3D dengan keterangan warna dan color bar
X, Y, Z = np.meshgrid(x_vals, y_vals, z_vals)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(X, Y, Z, c=result_vals, cmap='viridis')
ax.set_xlabel('Nilai $x_1$')
ax.set_ylabel('Nilai $x_2$')
ax.set_zlabel('Nilai $x_3$')
ax.set_title('Grafik Fungsi $f_m$ dengan $a_2$=0.3 dengan $x_1$+$x_2$+$x_3$=1')

# Tambahkan color bar sebagai keterangan warna
colorbar = plt.colorbar(scatter)
colorbar.set_label('Nilai Fungsi $f_m$')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(a, x, y):
    z = 1 - x - y
    return (a * x * y * z) / ((1 + x + z) * (1 + y))

# Inisialisasi array nilai x dan y
x_vals = np.linspace(0, 1, 100)
y_vals = np.linspace(0, 1, 100)

# Inisialisasi beberapa nilai konstanta a
a_vals = [0.3]

# Membuat plot
plt.figure(figsize=(12, 8))

for a in a_vals:
    # Membuat grafik 3D
    X, Y = np.meshgrid(x_vals, y_vals)
    Z = f(a, X, Y)

    # Hapus nilai z ketika x + y + z > 1
    Z[X + Y > 1] = np.nan

    # Plot contour
    plt.subplot(1, len(a_vals), a_vals.index(a) + 1)
    plt.contourf(X, Y, Z, levels=100, cmap='viridis')
    plt.colorbar()
    plt.xlabel('Nilai $x_1$')
    plt.ylabel('Nilai $x_2$')
    plt.title(f'Grafik Fungsi $f_m$ dengan $a_2$={a} dengan $x_1$ +$x_2$ + $x_3$ = 1')

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Definisikan fungsi f(x, y, z) dengan a sebagai konstanta
def f(a, x, y, z):
    return a * x * y * z / ((1 + x + z) * (1 + y))

# Rentang nilai x, y, z
x_vals = np.linspace(0, 1, 100)
y_vals = np.linspace(0, 1, 100)
z_vals = np.linspace(0, 1, 100)

# Nilai konstanta a
a = 0.3

# Inisialisasi variabel untuk menyimpan nilai maksimum dan koordinat titik maksimum
max_value = -np.inf
max_coord = None

# Loop untuk mencari nilai maksimum di setiap titik
for x in x_vals:
    for y in y_vals:
        for z in z_vals:
            # Berdasarkan batasan x + y + z = 1
            if x + y + z <= 1:
                value = f(a, x, y, z)
                # Perbarui nilai maksimum dan koordinat titik maksimum jika ditemukan nilai yang lebih besar
                if value > max_value:
                    max_value = value
                    max_coord = (x, y, z)

print("Nilai Maksimum Fungsi:", max_value)
print("Koordinat Titik Maksimum:", max_coord)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Definisikan fungsi f(x, y, z) dengan a sebagai konstanta
def f(a, x, y, z):
    return a * x * y * z / ((1 + x + z) * (1 + y))

# Definisikan fungsi g(x, y, z) dengan a sebagai konstanta
def g(a, x, y, z):
    return a * x * y * z / (1 + x + y + z)

# Rentang nilai x, y, z
x_vals = np.linspace(0, 1, 100)
y_vals = np.linspace(0, 1, 100)
z_vals = np.linspace(0, 1, 100)

# Inisialisasi array untuk menyimpan nilai fungsi f(x, y, z) dan g(x, y, z)
result_f_vals = np.zeros((100, 100, 100))
result_g_vals = np.zeros((100, 100, 100))

# Nilai konstanta a
a = 0.3

# Loop untuk menghitung nilai fungsi pada setiap titik
for i in range(len(x_vals)):
    for j in range(len(y_vals)):
        for k in range(len(z_vals)):
            x = x_vals[i]
            y = y_vals[j]
            z = z_vals[k]

            # Berdasarkan batasan x + y + z = 1
            if x + y + z <= 1:
                result_f_vals[i, j, k] = f(a, x, y, z)
                result_g_vals[i, j, k] = g(a, x, y, z)
            else:
                result_f_vals[i, j, k] = np.nan
                result_g_vals[i, j, k] = np.nan

# Plot grafik 3D fungsi f(x, y, z)
X, Y, Z = np.meshgrid(x_vals, y_vals, z_vals)
fig_f = plt.figure()
ax_f = fig_f.add_subplot(111, projection='3d')
f_plot = ax_f.scatter(X, Y, Z, c=result_f_vals, cmap='viridis')
ax_f.set_xlabel('Nilai x')
ax_f.set_ylabel('Nilai y')
ax_f.set_zlabel('Nilai z')
ax_f.set_title('Grafik Fungsi f(x, y, z) dengan a=0.3 pada x+y+z=1')
fig_f.colorbar(f_plot)

# Plot grafik 3D fungsi g(x, y, z)
fig_g = plt.figure()
ax_g = fig_g.add_subplot(111, projection='3d')
g_plot = ax_g.scatter(X, Y, Z, c=result_g_vals, cmap='viridis')
ax_g.set_xlabel('Nilai x')
ax_g.set_ylabel('Nilai y')
ax_g.set_zlabel('Nilai z')
ax_g.set_title('Grafik Fungsi g(x, y, z) dengan a=0.3 pada x+y+z=1')
fig_g.colorbar(g_plot)

plt.show()
```
