#**If Else and If Elif Else**

##**If Else**

```python
a = int(input("Enter the value of a:"));
b = int(input("Enter the value of b:"));
if a > b:
  print("a=", a,"is greater than b=",b)
if a == b:
  print("a=", a,"is equal to b=",b)
else:
  print("a=", a,"is less than b=",b)
```

##**If Elif Else**

```python
a = int(input("Enter the value of a:"));
b = int(input("Enter the value of b:"));
if a > b:
  print("a=", a,"is greater than b=",b)
elif a == b:
  print("a=", a,"is equal to b=",b)
else:
  print("a=", a,"is less than b=",b)
```

#**For and While**

```python
data = []
k = int(input("Enter the length of data list:"))
for i in range(k):
  data.append(int(input("Enter the data:")))
print(data)
```

```python
# checking whether the input number is a prime number or not
number = int(input("Enter the number greater than 2:"))
start = 2;
test = True
while start < number:
  if number%start == 0:
    start = number + 1
    test = False
  elif number%start != 0:
    start = start + 1

if test == False:
  print(number, "is not a prime number")
elif test == True:
  print(number, "is a prime number")
```

#**Function**
is a block of code than can be reused in a program

```python
# we create a factorial number program
def factorial(x):
  factorial = 1;
  for i in range (1,x+1):
    factorial = factorial*i
  return factorial
```

```python
x = int(input("Enter the number:"))
print(x,'!=', factorial(x))
```
